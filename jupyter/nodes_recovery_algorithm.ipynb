```python
import numpy as np
import pandas as pd
import operator

class VDsDistributer:
    def __init__(self,entries,capacityArray, requestsArray):
        self.entries=entries
        self.capacityArray=capacityArray
        self.requestsArray=requestsArray
        
    def recoveryDistribution(self):
        column_names = ["Max", "Chosen_N", "capacity", "requests"]
        df = pd.DataFrame(columns = column_names)
        df = df.append([{'capacity': ', '.join(map(str,capacityArray.tolist())), 'requests':', '.join(map(str,requestsArray.tolist()))}], ignore_index=True)

        #algorithm
        for i in entries:
            a = np.take(capacityArray, i-1)
            b = np.take(requestsArray, i-1)
            c = a/b
            cMaxIndex, cMax = max(enumerate(c), key=operator.itemgetter(1))
            for j in i:
                requestsArray [j-1] -= 1
            capacityArray[i[cMaxIndex] - 1] -= 1
            maxString = list(map(lambda x, y: x+ '/' +y, map(str,a), map(str,b)))
            df = df.append([{'Max':', '.join(maxString), 'Chosen_N': i[cMaxIndex], 'capacity': ', '.join(map(str,capacityArray.tolist())), 'requests':', '.join(map(str,requestsArray.tolist()))}], ignore_index=True)

        display(df)
```

```python
#parameters example
entries = np.array([[1,2,4],[1,3,4],[2,3,5],[1,3,4],[2,3,5],[1,2,5],[1,3,5],[1,3,4],[1,2,3],[1,2,5],[1,4,5],[3,4,5],[1,3,4],[1,2,4],[1,3,5],[1,3,4],[2,3,5],[1,3,4],[1,3,4],[1,4,5]])
capacityArray = np.array([4,2,5,8,3])
requestsArray = np.array([16,8,14,12,10])

_vdsDistributer = VDsDistributer(entries, capacityArray, requestsArray)
_vdsDistributer.recoveryDistribution()
```
